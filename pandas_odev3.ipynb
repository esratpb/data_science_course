```python
def teksinif(i):
    data=pd.read_excel("C:\\Users\\Userr\\Desktop\\datascience_odev3\\sinif_data\\py_mind.xlsx",sheet_name=i)
    data1=data.iloc[-3:,1]
    df=pd.DataFrame({"sinif_ismi":"py_mind","ogrenci_adi":data.columns[0],"dogru":data1.iloc[0],"yanlis":data1.iloc[1],"bos":data1.iloc[2]},index=[i])
    return df   
    
```
